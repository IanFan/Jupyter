### Syntax

```python
df = pd.DataFrame(
    {'a': [4, 5, 6],
     'b': [7, 8, 9],
     'c': [10, 11, 12]},
    index=[1, 2, 3])
print('df')
print(df)

df = pd.DataFrame(
    [[4, 7, 12],
     [5, 8, 11],
     [6, 9, 10]],
    index=[1, 2, 3],
    columns=['a', 'b', 'c']
)
print('df')
print(df)

df = pd.DataFrame(np.array(
    [[4, 7, 12],
     [5, 8, 11],
     [6, 9, 10]]), 
    index=list('123'), 
    columns=list('abc'))
print('df')
print(df)
```

### Reshape

```python
df1 = pd.DataFrame(
    [[4, 7, 12],
     [5, 8, 11],
     [6, 9, 10]],
    index=[1, 2, 3],
    columns=['a', 'b', 'c']
)

df2 = pd.DataFrame(
    [[13, 15, 17],
     [14, 16, 18]],
    index=[1, 5],
    columns=['a', 'b', 'c']
)

df3 = pd.DataFrame(
    [[19, 21],
     [20, 22]],
    index=[1, 2],
    columns=['d', 'e']
)

melt_df = pd.melt(df1)
print('melt')
print(melt_df)

melt_df['index'] = [1,2,3,1,2,3,1,2,3]
pivot_df = melt_df.pivot(index='index', columns='variable', values='value')
print('pivot')
print(pivot_df)

con_df = pd.concat([df1, df2])
print('concate')
print(con_df)

con_df2 = pd.concat([df1, df3], axis=1)
print('concate axis=1')
print(con_df2)

print('sort values')
print(con_df2.sort_values(['a', 'b', 'c'], ascending=False))

print('sort index')
print(con_df2.sort_index(ascending=False))

con_df2 = con_df2.rename(columns={'a':'A'})
print('rename')
print(con_df2)

reset_index_df = con_df2.reset_index()
print('reset index')
print(reset_index_df)

drop_df = reset_index_df.drop(columns=['b', 'c'])
print('drop')
print(drop_df)
```

### Sub Rows

```python
df = pd.DataFrame(
    {'a': [4, 4, 6],
     'b': [7, 7, 7],
     'c': [7, 7, 12]},
    index=[1, 3, 3])
print('df')
print(df)

select_rows_df = df[df['b'] > 7]
# select_rows_df = df1[1:2]
print('select rows')
print(select_rows_df)

# drop_duplicates_df = df.drop_duplicates(keep='first')
drop_duplicates_df = df.drop_duplicates('b', keep='last')
print('drop duplicates')
print(drop_duplicates_df)

print('head')
print(df.head(2))

print('tail')
print(df.tail(2))

print('sample frac')
print(df.sample(frac=0.5))

print('sample n')
print(df.sample(n=2))

print('loc')
print(df.loc[2:3])

print('iloc')
print(df.iloc[1:2])

print('n largest')
print(df.nlargest(2, 'c'))

print('n samllest')
print(df.nsmallest(2, 'c'))
```
