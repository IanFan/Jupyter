```python
n_1 = ['19', '26', '34', '78', '91', '62', '43', '87']
n_2 = ['16', '29', '38', '47', '61', '92', '83', '74']
n_3 = ['13', '27', '49', '68', '31', '72', '94', '86']
n_4 = ['14', '28', '39', '67', '41', '82', '93', '76']
n_5 = ['17', '23', '46', '89', '71', '32', '64', '98']
n_6 = ['12', '37', '48', '69', '21', '73', '84', '96']
n_7 = ['18', '24', '36', '79', '81', '42', '63', '97']
n_8 = ['01', '02', '03', '04', '05', '06', '07', '08', '09',
       '10', '20', '30', '40', '50', '60', '70', '80', '90',
       '11', '22', '33', '44', '55', '66', '77', '88', '99',
       '15', '25', '35', '45', '65', '75', '85', '95', '00',
       '51', '52', '53', '54', '55', '56', '57', '58', '59']
n_list = [n_1, n_2, n_3, n_4, n_5, n_6, n_7, n_8]
c_list = ['①', '②', '③', '④', '⑤', '⑥', '⑦', '㊀']

chakra_list = ['1', '2', '3', '4', '5', '6', '7', '8']

match_dict = {'13': '①+③', '15': '①+⑤', '16': '①+⑥', '34': '③+④', '36': '③+⑥',
              '25': '②+⑤', '26': '②+⑥', '56': '⑤+⑥', '134': '①+③+④', '256': '②+⑤+⑥',
              '17': '⑦+①', '27': '⑦+②', '37': '⑦+③', '47': '⑦+④', '57': '⑦+⑤', '67': '⑦+⑥'}

match_describe_dict = {'13': '帶財帶庫', '15': '談判高手', '16': '白手起家', '34': '富貴命', '36': '帶財帶庫',
                       '25': '小人不斷、玻璃心', '26': '少爺公主命', '56': '心直口快', '134': '施主', '256': '三兇數人',
                       '17': '興大業、完美主義', '27': '婚破、不婚、怪婚', '37': '主大財、自私、計較', '47': '主大名、憂鬱、躁鬱', '57': '血光、官符、花錢消災', '67': '大破財、合夥必敗'}

chakra_describe_dict = {'1': '海底輪，延年，命帶事業',
                        '2': '本我輪，六煞，命帶桃花',
                        '3': '胃輪，天醫，命帶財庫、文昌',
                        '4': '心輪，生氣，命帶貴人',
                        '5': '喉輪，禍害，命帶口舌、明小人',
                        '6': '眉心輪，絕命，冒險、聰明',
                        '7': '頂輪，五鬼，大好大壞',
                        '8': '變卦，伏位，不吉、不兇'}

address_describe_dict = {'1': '事業房，助事業、睡不飽', '3': '財庫房，平順、平淡', '4': '貴人房、熱鬧、空洞', '8': '小吉房，平順、依主人運',
                         '7': '大好大壞房，三兇潛藏、不穩、不利睡眠', '2': '六煞房，犯小人不利感情、孤單、沒動力', '5': '口舌房，不利貴人、爭吵、傳統', '6': '漏財房，不利投資、愛花錢、生活品質佳',
                         '27': '婚破房，不利婚姻、檔婚姻', '57': '官符房，犯官司、血光', '67': '大破財房，大破財、財不聚'}

match_list = ['13', '15', '16', '25', '26', '34', '36', '56', '134', '256', '17', '27', '37', '47', '57', '67']

reverse_dict = {'1': '2', '2': '1', '3': '6', '6': '3', '4': '5', '5': '4', '7': '8', '8': '7'}

alpha_num_dict = {'A': '01', 'B': '02', 'C': '03', 'D': '04', 'E': '05', 'F': '06', 'G': '07', 'H': '08', 'I': '09',
                  'J': '10', 'K': '11', 'L': '12', 'M': '13', 'N': '14', 'O': '15', 'P': '16', 'Q': '17', 'R': '18',
                  'S': '19', 'T': '20', 'U': '21', 'V': '22', 'W': '23', 'X': '24', 'Y': '25', 'Z': '26'}

foo_prf = 0


def output_list_2_str(l):
    # 輸出文字
    out_str = ''
    last_str = ''
    for i in l:
        if len(i) > 0:
            if last_str != i:
                if len(out_str) > 0:
                    out_str += '\n'
                out_str += i
            last_str = i
    return out_str


def get_example():
    global foo_prf
    foo = foo_prf
    o_ = '例如輸入: 390827,391008\n\n'
    o_ += message2chakra('390827,391008')
    foo_prf = foo
    return o_


def get_feedback():
    o_ = '豐盛幸福圈-西蒙師\n'
    o_ += 'https://www.facebook.com/MasterSimonn'
    return o_


def get_support():
    return '技術總監 IAN 范財銘\nemail: ianfan0704@gmail.com'


def num2chakra(num):
    if int(num) in [1, 2, 3, 4, 5, 6, 7, 8]:
        return c_list[int(num)-1]
    else:
        print('error num2chakra: ' + str(num))
        return num


def get_foo_prf():
    global foo_prf
    if foo_prf >= 2:
        foo_prf = 0
        return get_support()
    elif foo_prf >= 1:
        foo_prf += 1
        return get_feedback()
    foo_prf += 1
    return get_example()


def remove_mid_5(a):
    l_ = list(a)
    is_any_mid_5 = False
    for i in range(len(l_)):
        if i != 0 and i != len(l_) - 1:
            if l_[i] == '5':
                l_[i] = ''
                is_any_mid_5 = True
    if is_any_mid_5 is True:
        return ''.join(l_)
    else:
        return a


def is_match_num(a):
    is_number_match = False
    if len(a) == 3 or len(a) == 5:
        if len(a) == 3 and a[1] == '+' and re.match('^[1-8]+$', a[0] + a[2]):
            is_number_match = True
        elif len(a) == 5 and a[1] + a[3] == '++' and re.match('^[1-8]+$', a[0] + a[2] + a[4]):
            is_number_match = True
    return is_number_match


def match_info(a):
    list_ = []
    for i in a:
        if re.match('^[1-8]+$', i):
            list_.append(i)
    list_.sort()
    match_num = ''.join(list_)
    if match_num in match_list:
        match_chakra = match_dict[match_num]
        match_describe = match_describe_dict[match_num]
        des = match_chakra + '\n' + match_describe
        if match_num in address_describe_dict.keys():
            des = des + '\n' + address_describe_dict[match_num]
        return des
    else:
        return get_foo_prf()


def single_chakra_info(a):
    n_l = n_list[int(a) - 1]
    s_ = ''
    wrap = 9 if int(a) == 8 else 4
    for i in range(len(n_l)):
        n = n_l[i]
        if i > 0 and i % wrap == 0:
            s_ += '\n'
        elif len(s_) > 0:
            s_ += ', '
        s_ += n
    return s_


def single_chakra_describe(a):
    describe = chakra_describe_dict[a]
    return describe


def filter_alphabet_num(a):
    if not re.match('^[a-zA-Z0-9]+$', a):
        a_ = ''
        for i in a:
            if re.match('^[a-zA-Z0-9]+$', i):
                a_ += i
        a = a_
    return a


def alphabet2num(a):
    a = a.upper()
    for key in alpha_num_dict.keys():
        a = a.replace(key, alpha_num_dict[key])
    return a


def is_any_alphabet(a):
    if re.search('[a-zA-Z]', a):
        return True
    return False


def remove_month_0(a):
    is_change_month = False
    year = a[:-4]
    month = a[-4:-2]
    day = a[-2:]
    if 1 <= int(day) <= 31:
        if month in ['01', '02', '03', '04', '05', '06', '07', '08', '09']:
            month = month[-1:]
            is_change_month = True
    if is_change_month is True:
        a = ''.join(year + month + day)
    return a


def numbers2doubles(a):
    r = []
    n_two_digit = []
    l_ = list(a)
    for i in range(len(l_)):
        if i > 0:
            n = l_[i - 1] + l_[i]
            n_two_digit.append(n)
            is_correct = False
            for j in range(len(n_list)):
                n_l = n_list[j]
                if n in n_l:
                    is_correct = True
                    r.append(str(j))
                    break
            if is_correct is False:
                print('error numbers2doubles: ' + str(n))
    str_ = ''
    for i in n_two_digit:
        if len(str_) > 0:
            str_ += ', '
        str_ += i
    return r, str_


def chakras2words(r):
    r_copy = r.copy()
    for i in range(len(r_copy)):
        n = r_copy[i]
        r_copy[i] = c_list[int(n)]
    c_copy = ''.join(r_copy)
    return c_copy


def change_chakra_8(r):
    c_8 = str(8-1)
    last_n = c_8
    is_any_c8 = False
    for i in range(len(r)):
        n = r[i]
        if n != c_8:
            last_n = n
        if n == c_8 and last_n != c_8:
            is_any_c8 = True
            r[i] = last_n
    return is_any_c8, r


def is_birthday(a):
    if len(a) == 6 or len(a) == 7:
        return True
    return False


def change_reverse(r):
    count = 0
    last = ''
    should_reverse = False
    for i in r:
        if i != last:
            count = 0
        count += 1
        last = i
    if count >= 4:
        should_reverse = True
    reverse = ''
    if should_reverse:
        reverse = reverse_dict[str(int(last)+1)]
    return should_reverse, reverse


def chakras2nums(c):
#     print(c)
    is_valid = True
    if not re.match('^[1-8]+$', c):
        print('error chakras2nums 1-8: ' + c)
        is_valid = False

    if is_valid is False:
        c_ = ''
        for i in c:
            if re.match('^[1-8]+$', i):
                c_ += i
        c = c_
#         print(c)

    before_change_8 = []
    for i in c:
        before_change_8 += str(int(i)-1)
    is_changed_8, changed_8 = change_chakra_8(before_change_8)
    if is_changed_8:
        after_change_8 = []
        for i in changed_8:
            after_change_8 += str(int(i)+1)
        c = ''.join(after_change_8)

    last = ''
    for i in c:
        if i == '8' and len(last) > 0 and last != '8':
            print('error chakras2nums 8: ' + c)
            return c
        last = i

    nums = ''
    for i in range(len(c)):
        c_ = c[i]
        n = n_list[int(c_)-1]
        if len(nums) > 0:
            last_n = nums[-1]
            if i < len(c) - 1 and c_ == '8' and c[i + 1] != 8:
                next_c = c[i + 1]
                n_next = n_list[int(next_c) - 1]
                for j in range(len(n)):
                    is_got = False
                    for c_next_ in n_next:
                        if n[j][-1] == c_next_[0]:
                            nums += n[j][-1]
                            is_got = True
                            break
                    if is_got is True:
                        break
            else:
                for j in range(len(n)):
                    if n[j][0] == last_n:
                        nums += n[j][-1]
                        break
        else:
            nums += n[0]
    return nums


def is_valid_chakra_words(ws):
    is_valid = True
    l_ = c_list + ['+']
    for w in ws:
        if w not in l_:
            is_valid = False
            break
    return is_valid


def chakra_words2chakra_nums(ws):
    ns = ''
    for w in ws:
        for i in range(len(c_list)):
            if c_list[i] == w:
                n = chakra_list[i]
                ns += n
    return ns


def chakra_nums_remove_duplicate(ns):
    ns_ = ''
    for n in ns:
        if n not in ns_:
            ns_ += n
    return ns_

def chakra2match(chakras):
    match_key_list = []
    for match in match_dict.keys():
        is_match = True
        for m in match:
            if m not in chakras:
                is_match = False
                break
        if is_match:
            match_key_list.append(match)
    match_l = []
    for m in match_key_list:
        match_l.append(match_dict[m])
    return match_l


def results2match(results):
    match = ''
    if len(results) >= 2:
        is_valid = True
        chakra_words = ''
        for result in results:
            rs = result.split('\n')
            c_words = rs[-1]
            if is_valid_chakra_words(c_words) is False:
                is_valid = False
                break
            else:
                chakra_words += (c_words.replace('+', ''))
        if is_valid is True:
            chakra_nums = chakra_words2chakra_nums(chakra_words)
            matches = chakra2match(chakra_nums)
            s_ = ''
            for m in matches:
                if len(s_) > 0:
                    s_ += ', '
                s_ += m
            match += s_
    return match


def msg2chakra(msg):
    global foo_prf
    a = msg

    # 去掉空格
    a = a.strip()

    output = []
    # 對碰資訊
    if is_match_num(a):
        info = match_info(a)
        output.append(info)
        return output

    # 只留下英文與數字
    a = filter_alphabet_num(a)
    output.append(a)

    # 提示資訊
    if len(a) == 0:
        output.append(get_foo_prf())
        return output

    if len(a) == 1:
        # 一個英文字母
        if is_any_alphabet(a):
            foo_prf = 0
            output.append(a)
        # 一個易數
        elif a in chakra_list:
            foo_prf = 0
            output.clear()
            output.append(num2chakra(a))
            output.append(single_chakra_info(a))
            output.append(single_chakra_describe(a))
            if a in address_describe_dict.keys():
                output.append(address_describe_dict[a])
            return output
        elif a == '9':
            foo_prf = 0
            info = ''
            for i in range(len(n_list)):
                if i != 0:
                    info += '\n\n'
                info += c_list[i]
                info += '\n'
                info += single_chakra_info(i + 1)
            output.clear()
            output.append(info)
            return output
        # 提示資訊
        else:
            output.clear()
            output.append(get_foo_prf())
            return output

    # 正常的輸入
    foo_prf = 0
    output.append(a)

    is_any_alpha = is_any_alphabet(a)

    is_bday = is_birthday(a) if is_any_alpha is False else False

    # 英文字母轉成01~26
    if is_any_alpha is True:
        a = alphabet2num(a)
        output.append(a)
    # 去掉年月日的月的0
    elif is_bday:
        a = remove_month_0(a)
        output.append(a)

    # 去掉夾在中間的5，個數5以下則不去除
    if len(a) >= 5:
        a = remove_mid_5(a)
        output.append(a)

    # 一串數字換成兩個兩個的數字
    r, doubles = numbers2doubles(a)
    output.append(doubles)

    # 易數顯示
    chakra_words = chakras2words(r)
    output.append(chakra_words)

    # 伏位換成前面的易數
    is_changed_8, change_8 = change_chakra_8(r)
    if is_changed_8:
        chakra_words = chakras2words(change_8)
        output.append(chakra_words)

    # 生日連續四個同數字則反轉
    if is_bday:
        should_reverse, reverse_c = change_reverse(change_8)
        if should_reverse:
            chakra_words = chakras2words(r) + '+' + num2chakra(int(reverse_c))
            output.append(chakra_words)

    return output


def message2chakra(msg):
    global foo_prf

    results = []
    msgs = re.split(',|，| ', msg)
    for m in msgs:
        if len(m) > 0:
            output_list = msg2chakra(m)
            results.append(output_list_2_str(output_list))

    match = results2match(results)
    if len(match) > 0:
        results += [match]

    output = ''
    for r in results:
        if len(output) > 0:
            output += '\n\n'
        output += r

    return output
```

```python
def test_english_name(name):
    result = message2chakra(name)
    chakra_words = ''
    rs = result.split('\n')
    c_words = rs[-1]
    
    if is_valid_chakra_words(c_words) is False:
        print('error is_valid_chakra_words', name)
        
    chakra_words += (c_words.replace('+', ''))
    chakra_nums = chakra_words2chakra_nums(chakra_words)
    chakra_nums_simple = chakra_nums_remove_duplicate(chakra_nums)
    chakra_nums_simple = list(map(int, chakra_nums_simple))
    chakra_nums_simple.sort()
    chakra_simple_sort = ''.join(str(x) for x in chakra_nums_simple)
    return chakra_simple_sort

def test_chakra2num():
    print('\n[test_chakra2num]\n')
    print(message2chakra(chakras2nums('88183')))
    print('\n--\n')
    print(message2chakra(chakras2nums('90239')))
```
