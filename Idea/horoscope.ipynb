```python
n_0 = ['01', '02', '03', '04', '05', '06', '07', '08', '09',
           '10', '20', '30', '40', '50', '60', '70', '80', '90',
           '11', '22', '33', '44', '55', '66', '77', '88', '99',
           '15', '25', '35', '45', '65', '75', '85', '95', '00',
           '51', '52', '53', '54', '55', '56', '57', '58', '59']
# n_1 = ['19', '26', '34', '78', '87', '43', '62', '91']
# n_2 = ['16', '29', '38', '47', '74', '83', '92', '61']
# n_3 = ['13', '27', '49', '68', '86', '94', '72', '31']
# n_4 = ['14', '28', '39', '67', '76', '93', '82', '41']
# n_5 = ['17', '23', '46', '89', '98', '64', '32', '71']
# n_6 = ['12', '37', '48', '69', '96', '84', '73', '21']
# n_7 = ['18', '24', '36', '79', '97', '63', '42', '81']
n_1 = ['19', '26', '34', '78', '91', '62', '43', '87']
n_2 = ['16', '29', '38', '47', '61', '92', '83', '74']
n_3 = ['13', '27', '49', '68', '31', '72', '94', '86']
n_4 = ['14', '28', '39', '67', '41', '82', '93', '76']
n_5 = ['17', '23', '46', '89', '71', '32', '64', '98']
n_6 = ['12', '37', '48', '69', '21', '73', '84', '96']
n_7 = ['18', '24', '36', '79', '81', '42', '63', '97']
n_list = [n_0, n_1, n_2, n_3, n_4, n_5, n_6, n_7]
c_list = ['㊀', '①', '②', '③', '④', '⑤', '⑥', '⑦']

alpha_num_dict = {'A':'01','B':'02','C':'03','D':'04','E':'05','F':'06','G':'07','H':'08','I':'09','J':'10','K':'11','L':'12','M':'13','N':'14','O':'15','P':'16','Q':'17','R':'18','S':'19','T':'20','U':'21','V':'22','W':'23','X':'24','Y':'25','Z':'26'}

foo_prf = 0

def output_list_2_str(l):
    # 輸出文字
    out_str = ''
    for i in l:
        if len(out_str) > 0:
            out_str += '\n'
        out_str += i
    return out_str

def get_example():
    global foo_prf
    foo = foo_prf
    o_ = '例: 690704\n\n'
    o_ += msg_2_chakra_number('690704')
    foo_prf = foo
    return o_

def get_feedback():
    o_ = '豐盛幸福圈-西蒙師\n'
    o_ += 'https://www.facebook.com/MasterSimonn'
    return o_

def get_support():
    return '技術開發 Ian 范財銘\nemail: ianfan0704@gmail.com'

def msg_2_chakra_number(msg):
    output = []
    a = msg
    
    # 去掉空格
    a = a.strip()

    # 只留下英文與數字
    if not re.match('^[a-zA-Z0-9]+$', a):
        a_ = ''
        for i in a:
            if re.match('^[a-zA-Z0-9]+$', i):
                a_ += i
        a = a_
        output.append(a)
        
    if len(a) == 0 or len(a) == 1:
        if a in ['0', '1', '2', '3', '4', '5', '6', '7']:
            global foo_prf
            foo_prf = 0
            print('foo_prf1' , foo_prf)
            output.append(c_list[int(a)])
            n_l = n_list[int(a)]
            s_ = ''
            wrap = 9 if int(a)==0 else 4
            for i in range(len(n_l)):
                n = n_l[i]
                if i>0 and i%wrap==0:
                    s_+= '\n'
                elif len(s_)>0:
                    s_ += ', '
                s_ += n
            output.append(s_)
            return output_list_2_str(output)
        elif (re.search('[a-zA-Z]', a)):
            foo_prf = 0
            output.append(a)
        else:
            if len(a) == 0 and foo_prf == 0:
                foo_prf += 1
            if foo_prf >= 3:
                foo_prf = 0
                return get_support()
            elif foo_prf >= 2:
                foo_prf += 1
                return get_feedback()
            elif foo_prf >= 1:
                foo_prf += 1
                return get_example()
            foo_prf += 1
            return '☯ 0~7'
    else:
        foo_prf = 0
        output.append(a)
    
    # 英文字母換成01~26
    is_any_alpha = False
    if (re.search('[a-zA-Z]', a)):
        is_any_alpha = True
    if is_any_alpha == True:
        a = a.upper()
        for key in alpha_num_dict.keys():
            a = a.replace(key, alpha_num_dict[key])
        output.append(a)

    # 去掉年月日的月的0
    if is_any_alpha == False:
        if len(a) == 6 or len(a) == 7:
            is_change_month = False
            year = a[:-4]
            month = a[-4:-2]
            day = a[-2:]
            if int(day) >= 1 and int(day) <= 31:
                if month in ['01', '02', '03', '04', '05', '06', '07', '08', '09']:
                    month = month[-1:]
                    is_change_month = True
            if is_change_month == True:
                a = ''.join(year + month + day)
                output.append(a)

    # 去掉夾在中間的5
    l = list(a)
    is_any_mid_5 = False
    for i in range(len(l)):
        if i != 0 and i != len(l) - 1:
            if l[i] == '5':
                l[i] = ''
                is_any_mid_5 = True
    if is_any_mid_5 == True:
        a = ''.join(l)
        output.append(a)
    
    # 一串數字換成兩個兩個的數字
    r = []
    n_two_digit = []
    l = list(a)
    for i in range(len(l)):
        if i > 0:
            n = l[i - 1] + l[i]
            n_two_digit.append(n)
            is_match_num = False
            for j in range(len(n_list)):
                n_l = n_list[j]
                if n in n_l:
                    is_match_num = True
                    r.append(str(j))
                    break
            if is_match_num == False:
                print('error' + str(n))
    str_ = ''
    for i in n_two_digit:
        if len(str_) > 0:
            str_ += ', '
        str_ += i
    if (len(a) > 2):
        output.append(str_)

    # 兩倆數字換成易數
    r_copy = r.copy()
    for i in range(len(r_copy)):
        n = r_copy[i]
        r_copy[i] = c_list[int(n)]
    c_copy = ''.join(r_copy)
    output.append(c_copy)

    # 伏位換成前面的易數
    last_n = '0'
    is_any_c0 = False
    for i in range(len(r)):
        n = r[i]
        if n != '0':
            last_n = n
        if n == '0' and last_n != '0':
            is_any_c0 = True
            r[i] = last_n
    if is_any_c0 == True:
        for i in range(len(r)):
            n = r[i]
            r[i] = c_list[int(n)]
        c = ''.join(r)
        output.append(c)

    return output_list_2_str(output)
```
